# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# View Sample

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('/..../Tiktok Popular Songs 2022/TikTok_songs_2022.csv')

# Display the first few rows of the DataFrame to verify the data
print(df.head())
```

# General Information

```python
# Track names
track_names = df['track_name']

# Artist names
artist_names = df['artist_name']

# Albums
albums = df['album']

# Popularity scores of tracks and artists
track_popularity = df['track_pop']
artist_popularity = df['artist_pop']

# Display the information
print("Track Names:")
print(track_names)

print("\nArtist Names:")
print(artist_names)

print("\nAlbums:")
print(albums)

print("\nPopularity Scores of Tracks:")
print(track_popularity)

print("\nPopularity Scores of Artists:")
print(artist_popularity)
```

## Top Tracks and Artists:

```python
top_tracks = df.sort_values(by='track_pop', ascending=False).head(10)

# Identify top artists
top_artists = df.sort_values(by='artist_pop', ascending=False).head(10)

print("Top Tracks:")
print(top_tracks[['track_name', 'track_pop']])

print("\nTop Artists:")
print(top_artists[['artist_name', 'artist_pop']])
```

## Album Analysis:

```python
album_distribution = df['album'].value_counts()

print("Album Distribution:")
print(album_distribution)
```

## Attribute Correlation:

```python
correlation_matrix = df[['danceability', 'energy', 'loudness']].corr()

print("Correlation Matrix:")
print(correlation_matrix)
```

## Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of track popularity
plt.hist(df['track_pop'], bins=20, color='skyblue')
plt.xlabel('Track Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Track Popularity')
plt.show()

# Scatter plot of energy vs. loudness
plt.scatter(df['energy'], df['loudness'], color='green', alpha=0.5)
plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.title('Energy vs. Loudness')
plt.show()

# Heatmap of correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Sorting 

```python
# Sort tracks by popularity
sorted_tracks = df.sort_values(by='track_pop', ascending=False)

# Sort artists by popularity
sorted_artists = df.sort_values(by='artist_pop', ascending=False)

print("Sorted Tracks by Popularity:")
print(sorted_tracks[['track_name', 'track_pop']])

print("\nSorted Artists by Popularity:")
print(sorted_artists[['artist_name', 'artist_pop']])
```

## Top Tracks and Artists:

```python
# Top Tracks
top_tracks = sorted_tracks.head(10)

# Top Artists
top_artists = sorted_artists.drop_duplicates(subset=['artist_name']).head(10)

print("Top Tracks:")
print(top_tracks[['track_name', 'track_pop']])

print("\nTop Artists:")
print(top_artists[['artist_name', 'artist_pop']])
```

## Distribution Analysis:

```python
# Distribution of Track Popularity
plt.hist(df['track_pop'], bins=20, color='skyblue')
plt.xlabel('Track Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Track Popularity')
plt.show()

# Distribution of Artist Popularity
plt.hist(df['artist_pop'], bins=20, color='salmon')
plt.xlabel('Artist Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Artist Popularity')
plt.show()
```

# Attribute Analysis:

```python
# Calculate descriptive statistics
attribute_stats = df[['danceability', 'energy', 'loudness', 'mode', 'key']].describe()
print(attribute_stats)

# Create histograms for each attribute
df[['danceability', 'energy', 'loudness', 'mode', 'key']].hist(figsize=(12, 8), bins=20, color='skyblue', alpha=0.7)
plt.suptitle('Distribution of Track Attributes', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()


# Pairplot to visualize relationships between attributes
import seaborn as sns

sns.pairplot(df[['danceability', 'energy', 'loudness', 'mode', 'key']], diag_kind='kde', corner=True)
plt.suptitle('Pairwise Relationships between Track Attributes', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

### Comparative Analysis:

      Comparing Danceability:

            If we want to see how easy it is to dance to one song compared to another, we can look at their danceability scores. A higher danceability score means the song is better for dancing, while a lower score means it's not as dance-friendly.

      Difference in Energy:

            We can check the energy levels of songs to see if there's a difference between tracks from different albums or artists. A higher energy score means the song is more lively and energetic, while a lower score means it's more mellow. By comparing the average energy levels of songs from different sources, we can get an idea of the overall vibe associated with each.


      Variation in Loudness Levels:

            When it comes to how loud a song is, there's a connection with how popular it is. Generally, more popular songs tend to be louder. However, it's important to remember that this doesn't mean loudness causes popularity or vice versa. It's just an observation that popular songs often have higher loudness levels.


      Correlation Matrix:

            The correlation matrix helps us understand how danceability, energy, and loudness are related. Here's what it tells us:
                  Danceability is slightly related to energy and loudness, but not very strongly. It means songs with higher danceability might have slightly lower energy levels and slightly higher loudness, but it's not a big difference.
                  Energy and loudness are moderately related. This means songs with higher energy levels usually tend to be louder, which makes sense because energetic songs often have more powerful sound.


### Pattern Recognition:

      Spotting Trends:

            We'll look for common themes or trends in the data. For example, we might find that energetic songs tend to be more danceable.

            We'll see if attributes like danceability and energy are linked. If they are positively correlated, it means that when one goes up, the other tends to go up too. If they're negatively correlated, it means when one goes up, the other tends to go down.
